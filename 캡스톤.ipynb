```python
# ì—¬ê¸° íŒŒì¼ ë°”ê¿”ì„œ ì§„í–‰
# ë°ì´í„° ì •ì œ
result = []
with open("/content/drive/Shareddrives/22-2 á„á…¢á†¸á„‰á…³á„á…©á†«/á„†á…®á†«á„’á…¡á†¨á„Œá…¡á†¨á„‘á…®á†·.txt/ğŸ‘OstrÌŒe-sledovaneÌ-vlaky-_Hrabal_-Bohumil_-_z-lib.org_.txt") as text_file:
    count = 0
    for line in text_file:
        newline = re.sub(r'[^\w\s]','',line)
        for word in newline.split():
            word = word.lower()
            if word not in stopwords :
                result.append(word)
```

```python
# ëª…ì‚¬ë§Œ ë½‘ì•„ë‚´ê¸°
noun_df = df[df['tag'].str.contains('NN')]
noun_df['lemma']
```

```python
# lemmaì—ì„œ ë¶ˆí•„ìš”í•œ ë‹¨ì–´ ë½‘ì•„ë‚´ê¸°
pattern = re.compile('^[\w]+')
noun_df['clean_lemma'] = noun_df.loc[:, 'lemma'].apply(lambda text: pattern.sub(' ', text))
```

```python
noun_df['clean_lemma'] = noun_df['clean_lemma'].str.strip()
```

```python
noun_df['result'] = [a.replace(b, '') for a, b in zip(noun_df['lemma'], noun_df['clean_lemma'])]
noun_df[:50]
```
