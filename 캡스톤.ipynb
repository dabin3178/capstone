```python
# Î∂àÏö©Ïñ¥ Î∂àÎü¨Ïò§Í∏∞
stopwords = unidecode.unidecode(open('/content/drive/Shareddrives/22-2 ·Ñè·Ö¢·Ü∏·Ñâ·Ö≥·Ñê·Ö©·Ü´/stopwords-cs.txt').read())
```

## Ïó¨Í∏∞ÏÑú Î∂ÄÌÑ∞

```python
# Ïó¨Í∏∞ ÌååÏùº Î∞îÍøîÏÑú ÏßÑÌñâ
# Îç∞Ïù¥ÌÑ∞ Ï†ïÏ†ú
result = []
with open("/content/drive/Shareddrives/22-2 ·Ñè·Ö¢·Ü∏·Ñâ·Ö≥·Ñê·Ö©·Ü´/·ÑÜ·ÖÆ·Ü´·Ñí·Ö°·Ü®·Ñå·Ö°·Ü®·Ñë·ÖÆ·Ü∑.txt/üëçOstrÃåe-sledovaneÃÅ-vlaky-_Hrabal_-Bohumil_-_z-lib.org_.txt") as text_file:
    count = 0
    for line in text_file:
        newline = re.sub(r'[^\w\s]','',line)
        for word in newline.split():
            word = word.lower()
            if word not in stopwords :
                result.append(word)
```

```python
print(result)
```

```python
clean2 = ' '.join(str(s) for s in result)
print(clean2)
```

```python
text_tagged = list(tagger.tag(clean2))
pprint(text_tagged)
```

```python
# Î™ÖÏÇ¨Îßå ÎΩëÏïÑÎÇ¥Í∏∞
noun_df = df[df['tag'].str.contains('NN')]
noun_df['lemma']
```

```python
# lemmaÏóêÏÑú Î∂àÌïÑÏöîÌïú Îã®Ïñ¥ ÎΩëÏïÑÎÇ¥Í∏∞
pattern = re.compile('^[\w]+')
noun_df['clean_lemma'] = noun_df.loc[:, 'lemma'].apply(lambda text: pattern.sub(' ', text))
```

```python
noun_df['clean_lemma'] = noun_df['clean_lemma'].str.strip()
```

```python
noun_df['result'] = [a.replace(b, '') for a, b in zip(noun_df['lemma'], noun_df['clean_lemma'])]
noun_df[:50]
```

Ïñ∏ÎçîÎ∞î ÏßÄÏö∞Í∏∞

```python
noun_df['underscore'] = '_'
noun_df[:50]
```

```python
noun_df['result2'] = [a.replace(b, '') for a, b in zip(noun_df['result'], noun_df['underscore'])]
noun_df[:50]
```
